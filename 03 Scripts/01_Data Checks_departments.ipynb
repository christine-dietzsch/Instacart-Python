```python
# checking for mixed-typed data:
for col in df_deps.columns.tolist():
  weird = (df_deps[[col]].applymap(type) != df_deps[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_deps[weird]) > 0:
    print (col)
```
